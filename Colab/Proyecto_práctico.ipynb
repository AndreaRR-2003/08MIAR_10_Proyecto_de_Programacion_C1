````python
env_name = "Enduro-v0"
env = gym.make(env_name)

np.random.seed(123)
env.seed(123)

nb_actions = env.action_space.n

print(env.action_space)
print(env.observation_space)
````

````python
from tensorflow.keras.layers import Input, Dense, Flatten, Convolution2D, Permute
from tensorflow.keras.models import Model

input_layer = Input(shape=(WINDOW_LENGTH, INPUT_SHAPE[0], INPUT_SHAPE[1]))

x = Permute((2, 3, 1))(input_layer)

x = Convolution2D(32, kernel_size=(8, 8), strides=(4, 4), activation="relu")(x)
x = Convolution2D(64, kernel_size=(4, 4), strides=(2, 2), activation="relu")(x)
x = Convolution2D(64, kernel_size=(3, 3), strides=(1, 1), activation="relu")(x)

x = Flatten()(x)
x = Dense(512, activation="relu")(x)

output_layer = Dense(nb_actions, activation="linear")(x)

model = Model(inputs=input_layer, outputs=output_layer)

model.summary()
````

````python
from rl.core import Processor
from PIL import Image
import numpy as np

INPUT_SHAPE = (84, 84)

class AtariProcessor(Processor):
    def process_observation(self, observation):
        assert observation.ndim == 3

        img = Image.fromarray(observation)
        img = img.resize(INPUT_SHAPE).convert('L')

        processed_observation = np.array(img)

        assert processed_observation.shape == INPUT_SHAPE

        return processed_observation.astype('uint8')

    def process_state_batch(self, batch):
        return batch.astype('float32') / 255.

    def process_reward(self, reward):
        return np.clip(reward, -1., 1.)
````

````python
processor = AtariProcessor()
print(processor)
````

````python
import json

json_loads_original = json.loads

def json_loads_patched(obj, *args, **kwargs):
    if isinstance(obj, dict):
        return obj
    return json_loads_original(obj, *args, **kwargs)

json.loads = json_loads_patched
````

````python
import rl.util
import tensorflow.keras.models as keras_models

def clone_model_patched(model, custom_objects=None):
    cloned_model = keras_models.clone_model(model)
    cloned_model.set_weights(model.get_weights())
    return cloned_model

rl.util.clone_model = clone_model_patched
````

````python
dqn = DQNAgent(
    model=model,
    nb_actions=nb_actions,
    memory=memory,
    policy=policy,
    processor=AtariProcessor(),
    nb_steps_warmup=50000,
    gamma=0.99,
    target_model_update=10000,
    train_interval=4,
    delta_clip=1.0,
    enable_double_dqn=True
)
````

````python
INPUT_SHAPE = (84, 84)
WINDOW_LENGTH = 4
nb_actions = env.action_space.n
````

````python
def preprocess_observation(observation):
    img = Image.fromarray(observation)
    img = img.resize(INPUT_SHAPE).convert("L")
    img = np.array(img, dtype=np.float32) / 255.0
    return img
````

````python
def create_initial_state(observation):
    frame = preprocess_observation(observation)
    state = np.stack([frame] * WINDOW_LENGTH, axis=-1)
    return state
````

````python
def update_state(state, observation):
    frame = preprocess_observation(observation)
    state = np.append(state[:, :, 1:], np.expand_dims(frame, axis=-1), axis=-1)
    return state
````

````python
def build_model():
    inputs = keras.Input(shape=(84, 84, 4))

    x = layers.Conv2D(32, 8, strides=4, activation="relu")(inputs)
    x = layers.Conv2D(64, 4, strides=2, activation="relu")(x)
    x = layers.Conv2D(64, 3, strides=1, activation="relu")(x)

    x = layers.Flatten()(x)
    x = layers.Dense(512, activation="relu")(x)

    outputs = layers.Dense(nb_actions, activation="linear")(x)

    return keras.Model(inputs, outputs)
````

````python
from tensorflow import keras
from tensorflow.keras import layers
q_network = build_model()
target_network = build_model()

target_network.set_weights(q_network.get_weights())

q_network.summary()
````
